```python
# clone repo to get liver data
!git clone https://github.com/jveenland/TM10007_ML.git

import pandas as pd
from IPython.display import display

# General packages
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets as ds
import seaborn

# Classifiers
from sklearn import model_selection
from sklearn import metrics
```

```python
df = pd.read_csv('TM10007_ML/worcliver/Liver_radiomicFeatures.csv', index_col=0)
display(df)
```

```python
import numpy as np

# split train and set data set in 80/20
ind = np.arange(df.shape[0])
np.random.shuffle(ind)
X_train = df.iloc[ind[:int(0.8*df.shape[0])],:]
X_test = df.iloc[ind[int(0.8*df.shape[0]):],:]

#display(X_test)

benign_count = X_test['label'].value_counts()['benign']
malign_count = X_test['label'].value_counts()['malignant']
print('Number of benign:', benign_count)
print('Number of malignant:', malign_count)

X_test.to_csv('TM10007_ML/worcliver/test_set.csv')
```
